```python
import pandas as pd
#averaging the values
frames_all=pd.read_csv('emotions_all.csv',dtype={'CALM':float,'SAD':float,'HAPPY':float,'ANGRY':float,'SURPRISED':float,'CONFUSED':float,'DISGUSTED':float,'FEAR':float})
li_col_name=['CALM','SAD','HAPPY','ANGRY','SURPRISED','CONFUSED','DISGUSTED','FEAR', 'vid_name']
df_avg=pd.DataFrame(data=[],columns=li_col_name)
row=1
for img_num in range(9704,9756):
    vid_name="IMG_"+str(img_num)
    vid_frames=frames_all.loc[frames_all.vid_name==vid_name]
    if not vid_frames.empty:
        df_avg.loc[row,'vid_name']=vid_name
        for i in range(0,len(li_col_name)-1):
            df_avg.loc[row,li_col_name[i]]=vid_frames[li_col_name[i]].sum()/vid_frames[li_col_name[i]].size
        row+=1
        
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
#train test data split - 1
train_data=df_merge[0:29]
test_data=df_merge[29:]
train_x = train_data.drop(columns=['vid_name','file_name','label'],axis=1)
train_y = train_data['label']
test_x = test_data.drop(columns=['vid_name','file_name','label'],axis=1)
test_y = test_data['label']

#train test data split - 2
X=df_merge.drop(columns=['vid_name','file_name','label'],axis=1)
Y=df_merge['label']
train_x,test_x,train_y,test_y=train_test_split(X, Y, test_size=0.2, random_state=12)

#datatype conversion
for i in range(0,8):
    train_x[li_col_name[i]]=train_x[li_col_name[i]].astype(float)
    test_x[li_col_name[i]]=test_x[li_col_name[i]].astype(float)
```

```python
model = XGBClassifier(max_depth=4, learning_rate=0.01, n_estimators=250, min_child_weight=1)
 
# fit the model with the training data
model.fit(train_x,train_y)

# predict the target on the test dataset
predict_test = model.predict(test_x)
print('\nTarget on test data',predict_test) 
 
# Accuracy Score on test dataset
accuracy_test = accuracy_score(test_y,predict_test)
print('\naccuracy_score on test dataset : ', accuracy_test)
```

```python
#K-fold cross validation
from sklearn.model_selection import KFold
import numpy as np
kf = KFold(n_splits = 3, shuffle = True)
scores = []
for i in range(0,8):
    X[li_col_name[i]]=X[li_col_name[i]].astype(float)
    
for i in range(5):
    result = next(kf.split(X), None)
    x_train = X.iloc[result[0]]
    x_test = X.iloc[result[1]]
    y_train = Y.iloc[result[0]]
    y_test = Y.iloc[result[1]]
    kf_model = model.fit(x_train,y_train)
    predictions = kf_model.predict(x_test)
    scores.append(kf_model.score(x_test,y_test))
print('Scores from each Iteration: ', scores)
print('Average K-Fold Score : ' , np.mean(scores))
print('Standard Deviation : ',np.std(scores))
```

```python
def plot_roc_curve(fpr, tpr, name):
    plt.plot(fpr, tpr, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--', label='No Skill')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve for {name}')
    plt.legend()
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score

X_reduced=X
seed=17
Y2 = LabelEncoder().fit_transform(Y)

X2 = StandardScaler().fit_transform(X_reduced)

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X2, Y2, test_size = 0.20, 
                                                    random_state = 12)

models = []
models.append(('XGBoost',XGBClassifier(max_depth=4, learning_rate=0.01, n_estimators=350, min_child_weight=1)))
models.append(('Random Forest', RandomForestClassifier(n_estimators=700)))
models.append(('Logistic Regression', LogisticRegression(random_state=seed)))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('Gaussian Naive Bayes', GaussianNB()))
models.append(('SVM', SVC(random_state=seed)))

names=[]
results=[]
# 10-fold cross validation
for name, model in models:
    kfold = KFold(n_splits=10, random_state=seed)
    cv_results = cross_val_score(model, X_Train, Y_Train, cv=kfold, scoring="accuracy")
    results.append(cv_results)
    names.append(name)
    msg = "%s:\nAverage K-Fold score-%f\nStandard deviation-%f" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    model.fit(X_Train,Y_Train)
    print(confusion_matrix(Y_Test,model.predict(X_Test)))
    print(classification_report(Y_Test,model.predict(X_Test)))
    
#     auc=roc_auc_score(Y_Test,model.predict(X_Test))
#     print('AUC - ',auc)
#     fpr,tpr,_=roc_curve(Y_Test,model.predict(X_Test))
#     plot_roc_curve(fpr,tpr, name)
    
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, cross_val_score
seed=17
Y2 = LabelEncoder().fit_transform(Y)

X2 = StandardScaler().fit_transform(X)

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X2, Y2, test_size = 0.20, 
                                                    random_state = 12)

models = []
models.append(('Logistic Regression', LogisticRegression(random_state=seed)))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('Gaussian Naive Bayes', GaussianNB()))
models.append(('SVM', SVC(random_state=seed)))

names=[]
results=[]
# 10-fold cross validation
for name, model in models:
    kfold = KFold(n_splits=5, random_state=seed)
    cv_results = cross_val_score(model, X_Train, Y_Train, cv=kfold, scoring="accuracy")
    results.append(cv_results)
    names.append(name)
    msg = "%s:\nAverage K-Fold score-%f\nStandard deviation-%f" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    model.fit(X_Train,Y_Train)
    print(confusion_matrix(Y_Test,model.predict(X_Test)))
    print(classification_report(Y_Test,model.predict(X_Test)))

# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Machine Learning algorithm comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
